```python
import pandas as pd
vendas = pd.read_csv('data/cafelura.csv')
vendas.head()
```

```python
vendas['mes'] = pd.to_datetime(vendas['mes'], format='%Y/%m/%d')
print('Quantidade de linhas e colunas', vendas.shape)
print('Quantidade de dados nulos', vendas.isna().sum().sum())
```

```python
from plotar import plotar
%matplotlib inline
```

```python
plotar('Vendas da Cafelura de 2017 e 2018', 'Tempo', 'Vendas', 'mes', 'vendas', vendas)
```

```python
quantidade_de_dias_de_fds = pd.read_csv('data/dias_final_de_semana.csv')
quantidade_de_dias_de_fds.head()
```

```python
quantidade_de_dias_de_fds['quantidade_de_dias'].values
```

```python
vendas['vendas_normalizadas'] = vendas['vendas'] / quantidade_de_dias_de_fds['quantidade_de_dias'].values
vendas.head()
```

```python
plotar('Vendas normalizadas da Cafelura de 2017 a 2018', 'Tempo', 'Vendas normalizadas', 'mes', 'vendas_normalizadas', vendas)
```
