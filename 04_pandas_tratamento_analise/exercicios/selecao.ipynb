```python
import pandas as pd
alunos = pd.DataFrame({'Nome': ['Ary', 'Cátia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
                        'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
                        'Idade': [15, 27, 56, 32, 42, 21, 19, 35], 
                        'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6], 
                        'Aprovado': [True, False, False, True, True, True, False, False]}, 
                        columns = ['Nome', 'Idade', 'Sexo', 'Notas', 'Aprovado'])
alunos
```

```python
# Selecionar apenas os alunos aprovados
selecao = alunos['Aprovado']
alunos_aprovados = alunos[selecao]
alunos_aprovados
```

```python
# Selecionar apenas os alunas aprovadas
selecao = (alunos['Aprovado']) & (alunos['Sexo']=='F')
alunas_aprovadas = alunos[selecao]
alunas_aprovadas
```

```python
# Selecionar alunos com idade entre 10 e 20 anos ou com idade maior ou igual a 40 anos.
selecao = ((alunos['Idade'] > 10) & (alunos['Idade'] < 20)) | (alunos['Idade'] >= 40)
alunas_idades = alunos[selecao]
alunas_idades
```

```python
# Crie um DataFrame somente com os alunos reprovados e mantenha neste DataFrame apenas as colunas Nome, Sexo e Idade, nesta ordem.
selecao = alunos['Aprovado'] == False
alunos_reprovados_cols = alunos.loc[selecao, ['Nome','Sexo','Idade']]
alunos_reprovados_cols
```

```python
selecao = alunos['Aprovado'] == False
reprovados = alunos[['Nome', 'Sexo', 'Idade']][selecao]
reprovados
```
